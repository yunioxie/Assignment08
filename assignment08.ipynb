## 01  

Open Addressing  
장점 : 테이블의 빈 공간을 활용해서 공간을 효율적으로 사용할 수 있다.  
단점 : 적재율이 높아질수록 테이블의 성능은 낮아진다.  
Chaining  
장점 : 해시 테이블이 꽉 차도 값을 계속 추가할 수 있다.  
단점 : 연결 리스트를 사용해서 구현이 어렵다.

## 02-5  

답 : ![02-5](images/image02-5.jpg)  
풀이과정 : 키 값을 13으로 나눈 나머지 주소에 넣다가 33부터 충돌한다. 처음 충돌했을 때는 바로 다음 공간이 비었다면 그 공간에 키 값을 넣는다. 빈 공간이 아니라면 나머지에 2^2을 더해서 다시 13으로 나눈 후의 나머지 주소에 넣고 또 충돌하면 3^2을 더한다. 이 과정을 충돌하는 값마다 순서대로 반복하면 위의 모양이 나온다.

## 02-7  

답 : 효율성이 낮아진다.  
풀이과정 : 맨 앞에 삽입하면 O(1)인데 맨 뒤에 삽입하면 리스트를 끝까지 검색해야 하므로 길이가 n이면 O(n)이다. 그래서 효율이 떨어진다.  

## 02-8  

답 : 삽입이 효율성이 낮아지고 삭제는 변화가 없고 검색은 없거나 높아진다.  
풀이과정 : 삽입은 정렬된 상태에서 위치를 찾아야 하므로 O(n)으로 효율성이 낮아진다. 삭제는 노드를 찾아야 하므로 효율성에서 변화가 없다. 실패하는 검색은 정렬이 되지 않으면 끝까지 탐색해야 해서 O(n)인데 정렬이 되면 크기를 비교해서 금방 중단할 수 있기 때문에 개선이 되어 효율성이 높아진다. 성공적인 검색은 정렬 여부와 관게없이 계속 키 값을 찾아야 해서 효율성의 변화가 거의 없다.
